```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
female_data = pd.read_csv('ANSUR_II_FEMALE.csv', encoding='latin1')
male_data = pd.read_csv('ANSUR_II_MALE.csv', encoding='latin1')

female_data.head()
```

```python
male_data.head()
```

```python
concatenated_dataframes = pd.concat([female_data , male_data])
concatenated_dataframes
```

PreProcessing:

```python
# Preprocess dateset for converting weight to km
concatenated_dataframes['weightkg'] = concatenated_dataframes['weightkg'] /10
concatenated_dataframes['weightkg']
```

```python
# Preprocess dateset for converting stature(height) to cm
concatenated_dataframes['stature'] = concatenated_dataframes['stature'] / 10
concatenated_dataframes['stature']
```

```python
# convert Gender column (str) ----> to number 
# female = 0 
# male   = 1

concatenated_dataframes['Gender'] = concatenated_dataframes['Gender'].replace(["Female" , "Male"] , [ 0 , 1 ])
concatenated_dataframes['Gender']
```

```python
from matplotlib.colors import ListedColormap

plt.scatter(concatenated_dataframes['stature'] , concatenated_dataframes['weightkg'] , c=concatenated_dataframes['Gender']==1 , marker= "o" , alpha= 0.5)
plt.scatter(concatenated_dataframes['stature'] , concatenated_dataframes['weightkg'] , c=concatenated_dataframes['Gender']==0 , marker= "o" , alpha= 0.5)

plt.legend(['men', 'women'])
plt.title('women & men stature and weigths')
plt.xlabel('height(cm)')
plt.ylabel('weight(kg)')
```

```python
# showing heights for women and men on the same plot
women_height = concatenated_dataframes.query('Gender == 0')['stature']
men_height = concatenated_dataframes.query('Gender == 1')['stature']

plt.hist(men_height, color= 'purple', edgecolor= 'black', alpha=0.7)
plt.hist(women_height, color='red', edgecolor='black', alpha=0.5)
plt.title('women & men height')
plt.legend(['men', 'women'])
plt.xlabel('height(cm)')
plt.ylabel('person count')
```

```python
men_height_mean = concatenated_dataframes.query('Gender == 1')['stature'].mean()
men_height_mean
```

```python
# Splitting dataset to train and test datasets (%80 for train and %20 for test)

from sklearn.model_selection import train_test_split

X = concatenated_dataframes[['stature', 'weightkg']].reset_index()
X
```

```python
Y = concatenated_dataframes['Gender'].reset_index()
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=300)
```

```python
# len(X_train)=6068*0.8 = 4854
X_train = X_train[['stature', 'weightkg']] # 6068 * 0.8 = 4854
X_train
```

```python
# len(X_test) = 6068 * 0.2 = 1214
X_test = X_test[['stature', 'weightkg']] # 6068 * 0.2 = 1214 rows
X_test
```

```python
# fiting object oriented KNN algorithm on the train dataset

X_train = X_train[['stature', 'weightkg']]
X_train = np.array(X_train)
X_train
```

```python
X_test = X_test[['stature', 'weightkg']]
X_test = np.array(X_test)
X_test
```

```python
Y_train = Y_train['Gender']
Y_train = np.array(Y_train)
Y_train
```

```python
Y_test = Y_test['Gender']
Y_test = np.array(Y_test)
Y_test
```

```python
from knn import KNN

knn = KNN(7)
knn.fit(X_train, Y_train)
```

```python
per1 = np.array([160.5, 65.4])
per2 = np.array([171.3, 84.2])
per3 = np.array([153.0, 46.9])
per4 = np.array([180.2, 90.4])
per5 = np.array([170, 70])

list_persons = [per1, per2, per3, per4, per5]
OUTPUTS = knn.predict(list_persons)
OUTPUTS
```

```python
for output in OUTPUTS:
    if output == 0:
        print("woman")
    elif output ==1:
        print("man")
```

```python
# Evaluate your KNN algorithm on the test dataset
knn.evaluate(X_test, Y_test)
```

```python
# test accuracy 
k = 3
test_accuracy_list = []
k_list = []

for i in range(15) :

    k_list.append(k)
    knn = KNN(k)
    knn.fit(X_train , Y_train)
    acc = knn.evaluate(X_test , Y_test)
    k+=2
    
    test_accuracy_list.append(acc)

plt.plot( k_list , test_accuracy_list )
plt.xlabel(" k values ")
plt.ylabel("Test Accuracy")
plt.show()
```

```python
# Calculating confusion matrix for test dataset
Y_pred = knn.predict(X_test)
print(len(Y_pred))
```

```python
confusion_matrix = np.zeros((2, 2))

for i in range(1214):
    if Y_test[i] == 0 and Y_pred ==0:
        confusion_matrix[0][0] += 1
    elif Y_test[i] == 0 and Y_pred == 1:
        confusion_matrix[0][1] += 1
    elif Y_test[i] ==1 and Y_pred == 0:
        confusion_matrix[1][0] +=1
    elif Y_test[i] == 1 and Y_pred ==1:
        confusion_matrix[1][1] += 1
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(Y_test, Y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [False, True])
cm_display.plot()
plt.xlabel('predicted label')
plt.ylabel('True label')
plt.show()
```
