```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
data = pd.read_csv('Boston.csv')
data.head()
```

```python
data = data.rename(columns= {'mdev':'PRICE'})
```

```python
data.corr()
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(data.corr(), cbar=True, square= True, fmt='.1f', annot=True, annot_kws={'size':9})
```

```python
plt.scatter(data['nox'], data['indus'])
plt.show()
```

```python
x= np.array((data['tax'], data['nox'])).T
print(x)
y= np.array(data['indus'])
print(x.shape)
y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=40)

class LLS:
    def __init__(self):
        ...
    
    def fit(self, x_train, y_train):
        w= np.matmul(np.matmul(np.linalg.inv(np.matmul(x_train.T, x_train)), x_train.T), y_train)
        return w
    def predict(self, w, x_test):
        y_pred = x_test * w
        return y_pred

lls= LLS()
w= lls.fit(x_train, y_train)
w
```

```python
a, b = np.meshgrid(x_train[:,0], x_train[:,1])
plane = w[0] * a + w[1] * b


fig = plt.figure(dpi=100, facecolor='w', edgecolor='k')
ax= plt.axes(projection='3d')
ax.plot_surface(a, b, plane, alpha=0.2)
plt.scatter(x_train[:,0], x_train[:,1], y_train, label='data')
plt.legend()
plt.show()
```
